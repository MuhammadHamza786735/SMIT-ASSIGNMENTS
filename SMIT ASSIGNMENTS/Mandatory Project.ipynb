```python
class BankAccount:
    def __init__(self, name, balance=0.0):
        """Initialize a new bank account with a name and an optional initial balance (default 0.0)."""
        self.name = name
        self.balance = balance
        self.transactions = []  # List to store transactions

    def deposit(self, amount):
        """Deposit money into the account."""
        if amount < 0:
            print("Error: Deposit amount must be positive.")
            return
        if amount == 0:
            print("Error: Cannot deposit zero.")
            return
        self.balance += amount
        self.transactions.append(('Deposit', amount, self.balance))
        print(f"{self.name} deposited ${amount}. New balance: ${self.balance}.")

    def withdraw(self, amount):
        """Withdraw money from the account."""
        if amount < 0:
            print("Error: Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print(f"Error: Insufficient funds for withdrawal. Current balance is ${self.balance}.")
            return
        if self.balance - amount < 0:
            print(f"Error: Withdrawal would result in negative balance. Current balance is ${self.balance}.")
            return
        self.balance -= amount
        self.transactions.append(('Withdrawal', amount, self.balance))
        print(f"{self.name} withdrew ${amount}. New balance: ${self.balance}.")

    def check_balance(self):
        """Check and return the current balance."""
        return self.balance

    def print_transactions(self):
        """Print the account statement, showing all deposits and withdrawals."""
        print(f"Account statement for {self.name}:")
        if not self.transactions:
            print("No transactions have been made.")
        else:
            for transaction in self.transactions:
                transaction_type, amount, balance_after = transaction
                print(f"- {transaction_type}: ${amount}. New Balance: ${balance_after}")


# Function to create an account
def create_account(name, balance=0.0):
    """Create a new bank account."""
    return BankAccount(name, balance)


# Function to deposit money into an account
def deposit_money(account, amount):
    """Deposit a specific amount into the account."""
    account.deposit(amount)


# Function to withdraw money from the account
def withdraw_money(account, amount):
    """Withdraw a specific amount from the account."""
    account.withdraw(amount)


# Function to check the balance of the account
def check_balance(account):
    """Check and return the balance of the account."""
    return account.check_balance()


# Function to print the account statement
def print_statement(account):
    """Print all transactions and account details."""
    account.print_transactions()


# Example Usage:

# Step 1: Create an account for John Doe
john_account = create_account("John Doe")

# Step 2: Deposit money into the account
deposit_money(john_account, 500)  # John deposits $500

# Step 3: Withdraw money from the account
withdraw_money(john_account, 200)  # John withdraws $200

# Step 4: Check the balance
balance = check_balance(john_account)
print(f"Current balance: ${balance}")

# Step 5: Print the account statement
print_statement(john_account)
```

```python
def safe_divide():
    try:
        # Request input from the user
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        
        # Attempt to perform division
        result = num1 / num2
        
    except ValueError:
        # Handles case when input is not a valid number
        print("Error: Please enter valid numbers.")
    except ZeroDivisionError:
        # Handles division by zero
        print("Error: Cannot divide by zero.")
    except Exception as e:
        # Handles any other unexpected errors
        print(f"Unexpected error: {e}")
    else:
        # If no errors, print the result
        print(f"The result of {num1} divided by {num2} is: {result}")
    finally:
        print("Operation completed.")

# Call the function to execute
safe_divide()
```
